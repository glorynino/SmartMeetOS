# connecte to API

```python

import os
from dotenv import load_dotenv
from pathlib import Path
import requests

# chemin vers le fichier .env dans le dossier
dotenv_path = Path(".env/development")
load_dotenv(dotenv_path=dotenv_path)

API_KEY = os.getenv("SMS_TO_API_KEY")

def send_sms(to_number: str, message: str, sender_id: str = "Hackathon"):
    url = "https://api.sms.to/sms/send"
    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }
    data = {
        "to": to_number,
        "message": message,
        "sender_id": sender_id
    }

    response = requests.post(url, json=data, headers=headers)
    if response.status_code == 200:
        result = response.json()
        if result.get("success"):
            print(f"✅ SMS envoyé ! ID : {result.get('message_id')}")
        else:
            print(f"❌ Erreur : {result.get('error')}")
    else:
        print(f"❌ Échec HTTP {response.status_code} : {response.text}")

# test on envoie un numero
send_sms("+213xxxxxxxxxx", "depresso habs ma tetroli")




```
